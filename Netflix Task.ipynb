```python
#Extracting Months out of Date Added to find the best month for a producer to release content
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['month'] = df['date_added'].dt.month_name()
print(df[['date_added', 'month']].head())
```

```python
trend = df['month'].value_counts()
```

```python
plt.figure(figsize=(10, 5))
trend.sort_index().plot(kind='bar', color='blue', edgecolor='black')
plt.title("Number of Netflix Releases Per Month")
plt.xlabel("Month")
plt.ylabel("Number of Releases")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Combining IMDB Ratings With Dataset to find top Rated Movies on Netflix
netflix_df = pd.read_csv("netflix_content.csv")
imdb_df = pd.read_csv("title.ratings.tsv.csv")

print("Netflix 'show_id' sample:")
print(netflix_df['show_id'].head())
print("\nIMDb 'tconst' sample:")
print(imdb_df['tconst'].head())

netflix_df['code'] = netflix_df['show_id'].str.replace("s", "").str.zfill(7).apply(lambda x: 'tt' + x)

imdb_df['code'] = imdb_df['tconst'].str.zfill(8)

print("\nTransformed Netflix 'code' sample:")
print(netflix_df['code'].head())
print("\nTransformed IMDb 'code' sample:")
print(imdb_df['code'].head())

matching_codes = set(netflix_df['code']).intersection(set(imdb_df['code']))
print(f"\nMatching codes: {matching_codes}")

if matching_codes:
    merged_df = pd.merge(netflix_df, imdb_df, on="code", how="inner")
    merged_df.to_csv("merged_data.csv", index=False)
    print("\nMerged DataFrame:")
    print(merged_df.head())
else:
    print("\nNo matching codes found. Check if the code transformation works as expected.")
```

```python
merged_df
```

```python
af = merged_df.sort_values(['averageRating'], ascending = False)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_rate['title'], top_rate['averageRating'], color='gold')  
plt.title('Top 20 Movies', fontsize=16)
plt.xlabel('Movies', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()

# Show the plot
plt.show()
```
